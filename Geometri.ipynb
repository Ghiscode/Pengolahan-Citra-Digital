```python
img = cv2.imread('image.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

pts_src = np.float32([
    [300, 230],   # kiri atas
    [850,300],    # kanan atas
    [850, 600],   # kanan bawah
    [300, 610]    # kiri bawah
])

width, height = 800, 400
pts_dst = np.float32([
    [0, 0],
    [width-1, 0],
    [width-1, height-1],
    [0, height-1]
])

# Transformasi perspektif
M = cv2.getPerspectiveTransform(pts_src, pts_dst)
warped = cv2.warpPerspective(img_rgb, M, (width, height))

# Tampilkan hasil
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Gambar Asli')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(warped)
plt.title('Tampak Lurus')
plt.axis('off')

plt.show()
```
